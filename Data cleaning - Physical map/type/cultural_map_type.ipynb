```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#read the file
PublicData = pd.read_csv("Public art.csv")
PublicData
```

```python
PublicData['Type'] = PublicData['Type'].str.split('(').str[0]
PublicData['Type'] = PublicData['Type'].str.split('.').str[0]
PublicData['Type'] = PublicData['Type'].str.split(',').str[0]
```

```python
PublicData.drop(['Title / subject','Artist / designer','y','x','Date','Location','Dimensions', 'Designation','Collection','Notes','Images','Source','Material'], axis=1, inplace = True)

PublicData
```

```python
PublicData.to_csv('PublicDataType.csv')
```

```python
f = open(u'PublicDataType.csv','r',encoding='utf-8').read()
wordcloud = WordCloud(
        background_color="white", 
        width=1500,              
        height=960,              
        margin=10              
        ).generate(f)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file('PublicDataType.png')
```
